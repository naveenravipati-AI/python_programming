# Sets

1) Unordered & Unindexed collection of items.
2) Set elements are unique. Duplicate elements are not allowed.
3) Set elements are immutable(cannot be changed).
4) Set itself is mutable. we can add or remove items from it 

**Set Creation**

```python
myset = {1,2,3,4,5} # set of numbers
myset
```

```python
len(myset)  #length of the set
```

```python
my_set = {1,1,2,2,3,3,3,4,4,6,7,8,8}
my_set # Duplicate elements are not allowed.
```

```python
myset1 = {1.79,2.02,1.89,7.0912} #set of float numbers
myset1
```

```python
myset2 = {'Asif','John','Tyrion'} #set of strings
myset2
```

```python
myset3 = {10,20.2,"Hola",(1,2,3)} #mixed datatypes
myset3
```

```python
myset3 = {10,20.2,"Hola",[1,2,3]}  #set doesn't allow mutable items like lists
myset3
```

```python
myset4 = set() # creates an empty set
print(type(myset4))
```

```python
my_set1 = set(('one','two','three','four'))
my_set1
```

**Copy Set**

```python
myset = {'one','two','three','four','five','six','seven'}
myset
```

```python
myset1 = myset # create a new reference myset1
myset1
```

```python
id(myset), id(myset1) # The address of both myset and myset1 will be same
```

```python
my_set = myset.copy() # create copy of the set
my_set
```

```python
id(my_set) #The address of my_set will be different from myset because it is not refered but creating a copy
```

```python
myset.add('eight')
myset
```

```python
myset1 #myset1 will be also impacted as it is pointing to the same set
```

```python
my_set #copy of the set won't be impacted due to changes made on the original set
```

# Set Operation

**Union**

```python
A = {1,2,3,4,5}
B = {4,5,6,7,8}
C = {8,9,10}
```

```python
A | B  # union of A & B (all elements from both sets. NO DUPLICATES)
```

```python
A.union(B) #union of A & B
```

```python
A.union(B,C) #union of A,B and C
```

```python
"""
updates the set, calling the update() method with union of A,B & C.
For below example Set A  will be updated with union of A,B & C.
"""
A.update(B,C)
A
```

**intersection**

```python
A = {1,2,3,4,5}
B = {4,5,6,7,8}
A & B #Intersection of A and B(Common items in both sets)
```

```python
A.intersection(B) # Intersection A and B
```

```python
"""
for below example Set A will be updated with Intersection of A and B.
"""
A.intersection_update(B)
A
```

**Difference**

```python
A = {1,2,3,4,5}
B = {4,5,6,7,8}
```

```python
A - B  # Set of elements that are only in A but not in B
```

```python
A.difference(B) # Difference of sets
```

```python
B - A # set of elements that are only in B but not in A
```

```python
B.difference(A)
```

```python
"""
for below example Set B will be updated with difference of B and A.
"""
B.difference_update(A)
B
```

**Symmetric Difference**

```python
A = {1,2,3,4,5}
B = {4,5,6,7,8} 
```

```python
A ^ B # Symmetric difference(set of elements in A & B but not in both)
```

```python
A.symmetric_difference(B)  #symmetric difference of sets
```

```python
"""
for below example Set A will be updated with symmetric difference of A and B.
"""
A.symmetric_difference_update(B)
A
```

**Subset, Superset & Disjoint**

```python
A = {1,2,3,4,5,6,7,8,9}
B = {3,4,5,6,7,8}
C = {10,20,30,40,50}
```

```python
B.issubset(A) # Set B is said to be the subset of set A if all elements of B are in A.
```

```python
A.issuperset(B) # set A is said to be the superset of set B if all elements of B are in A
```

```python
C.isdisjoint(A) # Two sets are said to be disjoint sets if they have no common elements
```

```python
B.isdisjoint(A) # Two sets are said to be disjoint sets if they have no common elements
```
